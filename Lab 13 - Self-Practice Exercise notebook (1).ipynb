**Problem 3:** Display basic statistical metrics

**Problem 4:** Generate a correlation matrix to analyze the relationships between features

```python
import pandas as pd

# Sample dataset
data = {
    'Feature1': [1, 2, 3, 4, 5],
    'Feature2': [2, 4, 1, 3, 5],
    'Feature3': [5, 4, 3, 2, 1]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Generating the correlation matrix
correlation_matrix = df.corr()

print("Correlation Matrix:")
print(correlation_matrix)
```

**Problem 5:** Create a histogram plot to visualize the distribution of the data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Values': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plotting a histogram
plt.hist(df['Values'], bins=5, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Values')
plt.grid(True)
plt.show()
```

**Problem 6:** Create a bar, pie chart plot to visualization of the data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Categories': ['A', 'B', 'C', 'D'],
    'Values': [30, 50, 20, 70]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plotting a bar chart
plt.bar(df['Categories'], df['Values'], color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart of Categories')
plt.grid(axis='y')  # Adding gridlines for better readability
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Categories': ['A', 'B', 'C', 'D'],
    'Values': [30, 50, 20, 70]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plotting a pie chart
plt.pie(df['Values'], labels=df['Categories'], autopct='%1.1f%%', startangle=90, colors=['gold', 'lightgreen', 'lightcoral', 'lightskyblue'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Pie Chart of Categories')
plt.show()
```
