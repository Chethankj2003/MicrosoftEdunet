```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample financial data
data = {
    'Date': ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16'],
    'Open': [774.25, 776.030029, 779.309998, 779, 779.659973],
    'High': [776.065002, 778.710022, 782.070007, 780.47998, 779.659973],
    'Low': [769.5, 772.890015, 775.650024, 775.539978, 770.75],
    'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')

# Set 'Date' column as index
df.set_index('Date', inplace=True)

# Plotting the data
plt.figure(figsize=(10, 6))

for column in df.columns:
    plt.plot(df.index, df[column], label=column)
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.figure(figsize=(10, 6))  # Set the figure size

# Create a bar chart
plt.bar(languages, popularity, color='blue')

# Title and labels
plt.title('Popularity of Programming Languages worldwide,oct 2017 compared to a year ago')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.grid(True)
# Display the bar chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Creating a pie chart
plt.figure(figsize=(8, 6))
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140)

# Adding a title
plt.title('Popularity of Programming Languages')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for X and Y axes
np.random.seed(42)  # Setting seed for reproducibility
x = np.random.rand(100)  # Generating 100 random numbers for X
y = np.random.rand(100)  # Generating 100 random numbers for Y

# Creating a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=50, alpha=0.7, c='red', edgecolors='black')

# Adding labels and title
plt.title('Random Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.tight_layout()

plt.show()
```
